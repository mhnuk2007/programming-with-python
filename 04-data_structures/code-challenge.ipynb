```python
my_str = 'AAABBBAAAAC'
def encodeString(stringVal):
    charList = []
    prevchar = stringVal[0]
    count = 0
    for char in stringVal:
        if char != prevchar:
            charList.append((prevchar, count))
            prevchar = char
            count = 0
        prevchar = char
        count += 1
    charList.append((prevchar, count))
    return charList

encodeString(my_str)
```

```python
my_list = [('A', 3), ('B', 3), ('A', 4), ('C', 1)]

def decodeString(encodedList):
    charList = []
    for (k, v) in encodedList:
        charList.append(k*v)

    return ''.join(charList)
    
decodeString(my_list)
        
```

```python
my_list = [('A', 3), ('B', 3), ('A', 4), ('C', 1)]
def decodeString2(encodedList):
    return ''.join(k*v for k,v in encodedList)

decodeString2(my_list)
```
