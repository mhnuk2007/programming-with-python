```python
def allPrimesUpTo(num):
    # Your code goes here.
    primes = []
    for number in range(2, num):
        for factor in range (2, int(number**0.5) + 1):
            if number % factor == 0:
                break
        else:
            primes.append(number)
    return primes
```
