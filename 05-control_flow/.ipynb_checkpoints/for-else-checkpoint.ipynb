## Find Primes

```python
primes = []
for number in range(2, 100):
    for factor in range (2, int(number**0.5) + 1):
        if number % factor == 0:
            break
    else:
        primes.append(number)
print(primes)
```

```python
for number in range(2, 100):
    found_factor = False
    for factor in range(2, int(number**0.5) + 1):
        if number % factor == 0:
            found_factor = True
            break
    if not found_factor:
        print (f'{number} is prime')
```

```python
primes_found = [2]
for number in range(2, 100):
    for factor in primes_found:
        if number % factor == 0:
            break
    else:
        primes_found.append(number)
print(primes_found)
```
